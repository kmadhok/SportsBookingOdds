```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
# Send a GET request to the covers.com NBA betting odds page
url = "https://www.covers.com/sport/basketball/nba/odds"
response = requests.get(url)

# Parse the HTML response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
game_rows_teams = soup.find_all("tr", class_="covers-CoversComponents-fixedColumnRow oddsGameRow")
game_rows = soup.find_all("tr", class_="covers-CoversOdds-mainTR oddsGameRow")

data = []
for row_teams, row_odds in zip(game_rows_teams, game_rows):
    away_team = row_teams.select_one('.__teams.__awaiting .__away img')
    if away_team:
        away_team = away_team['alt']
    home_team = row_teams.select_one('.__teams.__awaiting .__home img')
    if home_team:
        home_team = home_team['alt']
    odds_div = row_odds.find('div', class_='__bookOdds covers-Covers-Odds-withNoBorder')
    if odds_div is not None:
        try:
            away_odds = odds_div.find('div', class_='__awayOdds').find('div', class_='American').text.strip()
            home_odds = odds_div.find('div', class_='__homeOdds').find('div', class_='American').text.strip()
            bookmaker = row_odds.find('td', {'class': ['covers-CoversMatchups-centerAlignHelper', 'covers-CoversOdss-oddsTd', 'liveOddsCell']})['data-book']
            data.append({'Bookmaker': bookmaker, 'Away Team': away_team, 'Home Team': home_team, 'Away Odds': away_odds, 'Home Odds': home_odds})
        except AttributeError:
            pass

df = pd.DataFrame(data)
df
```
